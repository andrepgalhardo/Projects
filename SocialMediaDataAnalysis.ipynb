## Step 1: Importing Required Libraries

As the name suggests, the first step is to import all the necessary libraries that will be used in the project. In this case, we need pandas, numpy, matplotlib, seaborn, and random libraries.

Pandas is a library used for data manipulation and analysis. Numpy is a library used for numerical computations. Matplotlib is a library used for data visualization. Seaborn is a library used for statistical data visualization. Random is a library used to generate random numbers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import random
```

## 2- Create dataset

```python
social_media = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']
data = {'Date': pd.date_range('2021-01-01', periods=500), 'Category': [random.choice(social_media) for _ in range(500)], 'Likes': np.random.randint(0, 10000, size=500)}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.sample(30, random_state=100)
```

```python
df['Category'].value_counts()
```

## 3- Data Cleaning and Transforming

```python
df.info()
```

```python
df.isnull().any()
```

```python
df['WeekOfDay'] = df['Date'].dt.dayofweek
```

```python
df['Month'] = df['Date'].dt.month
```

```python
df.sample(20, random_state=100)
```

## 4 - Data Analysis

```python
sns.pairplot(df)
```

```python
df
```

```python
df.describe()
```

```python
df.describe(include = object)
```

```python
df.corr()
```

```python
sns.set_style('whitegrid')
f, ax = plt.subplots(figsize=(15,6))
ax = sns.lineplot(x='Date', y='Likes', data=df)
ax.set_xticklabels(labels = df['Date'], rotation=90)
plt.title("Likes x Time") 
plt.ylabel("Likes") 
plt.xlabel("Time")
monthyearFmt = mdates.DateFormatter('%Y-%m')
ax.xaxis.set_major_formatter(monthyearFmt)
sns.set_context("notebook", font_scale=1.0)
```

```python
f, ax = plt.subplots(figsize=(9,10))
plt.title("Likes x Category") 
plt.ylabel("Likes") 
plt.xlabel("Category") 
sns.set_context("notebook", font_scale=1.3)
sns.boxplot(x=df['Category'], y=df['Likes'])
```

```python
f, ax = plt.subplots(figsize=(9,7))
plt.title("Likes x Category") 
plt.ylabel("Likes") 
plt.xlabel("Category") 
sns.set_context("notebook", font_scale=1.3)
sns.barplot(x='Category', y='Likes', data=df, ci=False)
```

```python
df2 = df[['Likes', 'Category']]
```

```python
print('The average number of likes for category Food is', df2[df2['Category'] == 'Food'].mean())
print('The average number of likes for category Family is', df2[df2['Category'] == 'Family'].mean())
print('The average number of likes for category Culture is', df2[df2['Category'] == 'Culture'].mean())
print('The average number of likes for category Travel is', df2[df2['Category'] == 'Travel'].mean())
print('The average number of likes for category Fashion is', df2[df2['Category'] == 'Fashion'].mean())
print('The average number of likes for category Health is', df2[df2['Category'] == 'Health'].mean())
print('The average number of likes for category Fitness is', df2[df2['Category'] == 'Fitness'].mean())
print('The average number of likes for category Music is', df2[df2['Category'] == 'Music'].mean())
```

## 6 - Final Report

The data shown above was generated randomly and therefore follows a very close pattern. There are no major discrepancies between the categories, except between the 'Music' and 'Food' categories on the one hand and 'Travel' on the other hand with a more pronounced average.
